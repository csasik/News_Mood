```python
cnn = results_df.loc[results_df['Username'] == '@CNN',:]
bbc = results_df.loc[results_df['Username'] == '@BBC',:]
fox = results_df.loc[results_df['Username'] == '@Fox',:]
cbs = results_df.loc[results_df['Username'] == '@CBS',:]
nyt = results_df.loc[results_df['Username'] == '@NewYorktimes',:]
nyt.head()
```

```python
# Create plot
x_vals = bbc["Tweets Ago"]

fig, ax = plt.subplots()
ax.scatter(bbc["Tweets Ago"],bbc["Compound Score"],marker="o", alpha=0.8, color='orange', label='BBC')
ax.scatter(cnn["Tweets Ago"],cnn["Compound Score"],marker="o", alpha=0.8, color='red',label='CNN')
ax.scatter(fox["Tweets Ago"],fox["Compound Score"],marker="o", alpha=0.8, color='blue',label='Fox')
ax.scatter(cbs["Tweets Ago"],cbs["Compound Score"],marker="o", alpha=0.8, color='green',label='CBS')
ax.scatter(nyt["Tweets Ago"],nyt["Compound Score"],marker="o", alpha=0.8, color='violet',label='New York Times')

# # Incorporate the other graph properties
now = datetime.now()
now = now.strftime("%m/%d/%Y")
ax.legend(loc='right')
plt.title(f"Sentiment Analysis of Media Tweets ({now})")
plt.xlim([x_vals.max(),x_vals.min()]) #Bonus
plt.ylabel("Tweet Polarity")
plt.xlabel("Tweets Ago")
plt.savefig("Sentiment Analysis of Media Tweets.png")
plt.show()
```
